# Numpy Project :- Student Result Analysis

##### **Description About Data**
There is a 'students_marks.csv' file which contains data of 100 students.
There are 100 rows where every row representing a student data. 
There are four columns where first column contains roll number and 
next three columns contains marks of maths, physics and chemistry.

**Importing the numpy library**

```python
import numpy as np
```

**Read the data from csv file into numpy array**

```python
students_data=np.loadtxt('students_marks.csv',delimiter=',')
```

**Find the shape of data**

```python
students_data.shape
(100,4)
```

**Display first 5 rows of data**

```python
print(students_data[:5])
```

**Find the total of marks of each student**

```python
print(students_data[:,1:].sum(axis=1))
                         
```

**Add a new column that will store the total of marks of each student**

```python
total_marks=students_data[:,1:].sum(axis=1)
print(total_marks)
```

```python
total_marks=total_marks.reshape(-1,1)
print(total_marks[:5])
```

**Display first 5 rows of data with newly added column for total of marks**

```python
students_data=np.concatenate((students_data,total_marks),axis=1) #Broadcasting
students_data[:5]
```

**Display last 5 rows of data with newly added column for total of marks**

```python
students_data[-5:]
```

**Find the average of marks of each student**

```python
average_marks=(students_data[:,-1]/3).round(2)
print(average_marks)
```

**Add a new column that will store the average of marks of each student**

```python
average_marks=average_marks.reshape(-1,1)
students_data=np.concatenate((students_data,average_marks),axis=1)
```

**Display first 5 rows of data with newly added column for average of marks**

```python
print(students_data[:5])
```

**Display first 5 rows of data with newly added column for average of marks**

```python
print(students_data[-5:])
```

**Find the average of marks of maths subject**

```python
average_marks_math = students_data[:, 1].mean().round(2)
print(f"Average of marks of maths subject:{average_marks_math}")
```

**Find the average of marks of physics subject**

```python
average_marks_physics = students_data[:, 2].mean().round(2)
print(f"Average of marks of Physics subject:{average_marks_physics}")
```

**Find the average of marks of chemistry subject**

```python
average_marks_chemistry = students_data[:, 3].mean().round(2)
print(f"Average of marks of Chemistry subject:{average_marks_chemistry}")
```

**Find the highest marks in maths subject**

```python
highest_mark_math = round(students_data[:, 1].max())
print(f"Highest marks in maths subject:{highest_mark_math}")
```

**Find the highest marks in physics subject**

```python
highest_mark_physics = round(students_data[:, 2].max())
print(f"Highest marks in Physics subject:{highest_mark_physics}")
```

**Find the highest marks in chemistry subject**

```python
highest_mark_chemistry = round(students_data[:, 3].max())
print(f"Highest marks in Chemistry subject:{highest_mark_chemistry}")
```

**Find the lowest marks in maths subject**

```python
lowest_mark_math = round(students_data[:, 1].min())
print(f"Lowest marks in Maths subject:{lowest_mark_math }")
```

**Find the lowest marks in physics subject**

```python
lowest_mark_physics = round(students_data[:, 2].min())
print(f"Lowest marks in Physics subject:{lowest_mark_physics}")
```

**Find the lowest marks in chemistry subject**

```python
lowest_mark_chemistry = round(students_data[:, 3].min())
print(f"Lowest marks in Chemistry subject:{lowest_mark_chemistry}")
```

**Count the number of students who have scored more than or equal to 90 marks in maths subject**

```python
Student_count=(students_data[:, 1]>=90).sum()
print(f"Student scored >=90 in Maths:{Student_count}")
```

**Count the number of students who have scored more than or equal to 90 marks in physics subject**

```python
Student_count=(students_data[:, 2]>=90).sum()
print(f"Student scored >=90 in Physics:{Student_count}")
```

**Count the number of students who have scored more than or equal to 90 marks in Chemistry subject**

```python
Student_count=(students_data[:, 3]>=90).sum()
print(f"Student scored >=90 in Chemistry:{Student_count}")
```

**Count the number of students who have scored average marks more than or equal to 60**

```python
Student_count=(average_marks>=60).sum()
print(f"Number of Student Average Marks >=60: {Student_count}")
```

**Count the number of students who have scored average marks less than 50**

```python
print(f'''Number of students who have scored average marks less than 50 :
{(average_marks<50).sum()}''')
```

**Count the number of students who have scored average marks between 80 and 90**

```python
print(f'''Number of students who have scored average marks >80 and<90  :
{((average_marks>=80) & (average_marks<=90)).sum()}''')
```

**Count the number of students who have scored marks more than or equal to 90 in all subjects**

```python
print(f'''Number of students who have scored  marks >=90 in all subject  :
{((students_data[:,1] >= 90) & 
                 (students_data[:,2] >= 90) & 
                 (students_data[:,3] >= 90)).sum()
}''')
```

**Count the number of students who have scored marks more than or equal to 90 in any subject**

```python
print(f'''Number of students who have scored marks >=90 in any subject :
{((students_data[:,1] >= 90) | 
                 (students_data[:,2] >= 90) | 
                 (students_data[:,3] >= 90)).sum()
}''')
```

**Save the Students Result Analysis Data to new 'students_result.csv' file**

```python
np.savetxt(
    "students_result.csv",
    students_data,
    delimiter=",",
    header="Roll_Number,Math,Physics,Chemistry,Total_Marks,Average_Marks",
    fmt="%d,%d,%d,%d,%d,%.2f",
    comments=""
)
print("Data saved to students_result.csv")


```
